## 2. Undersampling to balance the classes

```python
y = df['Outcome']
X = df.drop(columns=['Outcome'])
```

```python
class_0 = df[df['Outcome'] == 0]
class_1 = df[df['Outcome'] == 1]
```

```python
minority_examples_count = len(class_1)
class_0_undersampled = class_0.sample(n=minority_examples_count, random_state=42)
```

```python
# Combine the two undersampled classes into a new DataFrame
undersampled_df = pd.concat([class_0_undersampled, class_1], axis=0)
```

```python
clase_0_count = undersampled_df['Outcome'].value_counts()[0]
clase_1_count = undersampled_df['Outcome'].value_counts()[1]

# Crear un gráfico de barras
plt.figure(figsize=(8, 6))
plt.bar(['Clase 0', 'Clase 1'], [clase_0_count, clase_1_count])
plt.xlabel('Clase')
plt.ylabel('Número de veces')
plt.title('Distribución de clases en la columna "Outcome"')
plt.show()
```

Ahora ya estan balanceadas!!!

```python
y = undersampled_df['Outcome']
X = undersampled_df.drop(columns=['Outcome'])
```

## 3. Split into train, test and validation

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=1) # 0.25 x 0.8 = 0.2
```

## 3. Build and compile Model

```python
from tensorflow.keras import Sequential, layers
from tensorflow.keras.layers import Dense, Dropout
```

```python
model = Sequential()
model.add(layers.Dense(units=512, activation='relu', input_shape=(X_train.shape[1],)))
model.add(layers.Dropout(0.05))
model.add(layers.Dense(units=256, activation='relu'))
model.add(layers.Dropout(0.05))
model.add(layers.Dense(units=128, activation='relu'))
model.add(layers.Dropout(0.05))
model.add(layers.Dense(units=64, activation='relu'))
model.add(layers.Dropout(0.05))
model.add(layers.Dense(units=32, activation='relu'))
model.add(layers.Dropout(0.05))
model.add(layers.Dense(units=1, activation='sigmoid')) 
```

```python
model.compile('Adam', loss='BinaryCrossentropy', metrics='accuracy')
```

```python
model.summary()
```

```python
hist = model.fit(X_train, y_train, epochs=50, validation_data=(X_val, y_val))
```

```python
plt.title('Loss')
plt.plot(hist.history['loss'], 'r')
plt.plot(hist.history['val_loss'], 'b')
plt.show()
```

```python
plt.title('Accuracy')
plt.plot(hist.history['accuracy'], 'r')
plt.plot(hist.history['val_accuracy'], 'b')
plt.show()
```

## 4. Make predictions

```python
#prediction
y_hat = model.predict(X_test)
y_hat = [0 if val <0.5 else 1 for val in y_hat]
```

```python
#Calculate accuracy of the prediction
accuracy_score(y_test, y_hat)
```

0.5% increase in accuracy just by balancing the classes
